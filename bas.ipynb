## Import Modules

```python
import pandas as pd
import altair as alt
```

## Clean data

```python
url = "https://www.espn.com/nba/player/gamelog/_/id/3112335/nikola-jokic"
df = pd.read_html(url)
jan = df[1].drop(columns=["Date","OPP","Result"])

jan = jan.drop(3).drop(17)
jan
```

```python
feb = df[0].drop(columns=["Date","OPP","Result"])

feb = feb.drop(7)
feb
```

```python
this_year = pd.concat([jan,feb])
this_year = this_year.reset_index().drop(columns=['index'])
this_year = this_year.rename(columns={"3P%":"Three_Percentage","3PT":"Three_PT","FG%":"FG_Percentage"})
```

```python
a = this_year.drop(columns=['FG','Three_PT','FT']).astype(float)
a
```

Question1: Will his FG% drop if he plays longer time?

```python

alt.Chart(a).mark_circle().encode(
    x=alt.X('MIN'),
    y=alt.Y('FG_Percentage')
)
```

According to the chart I made, we can see playing for a longer time does not make his percentage drop.

Answer1: He can play consistently.

## Make strings to float

```python
df1 =(this_year.FG
    .str.split('-',expand=True)
    .rename(columns = {0: 'Make', 1: 'Shoot'})
    .astype(float))

df1
```

```python
df2 = (this_year.FT
    .str.split('-',expand=True)
    .rename(columns = {0: 'FT_Make', 1: 'FT_Shoot'})
    .astype(float))

df2
```

```python
df3 = (this_year.Three_PT
    .str.split('-',expand=True)
    .rename(columns = {0: 'Three_Make', 1: 'Three_Shoot'})
    .astype(float))

df3
```

```python
re = pd.concat([df1,df2,df3],axis=1)
re
```

```python
re.mean()
```

```python
# When he shoot less than 50% 
a.query("FG_Percentage < 50").mean()
```
